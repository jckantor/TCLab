# Investigate why notebook and inline backends have different figure sizes.

The following two cells demonstrate the need to match `figsize` and `fig.dpi` to get plots of the same size with different backends.

## Restart Kernal before running

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

%matplotlib inline

def p():
    print(mpl.get_backend())
    print(mpl.rcParams['savefig.dpi'])
    print(mpl.rcParams['figure.dpi'])

    t = np.linspace(0,10,300)
    fig = plt.figure(figsize=(6,4), dpi=100)
    print("fig.dpi =", fig.dpi)
    print("fig size =", fig.get_size_inches())
    size = fig.get_size_inches()*fig.dpi # size in pixels
    print(size)
    plt.plot(t, np.sin(t))
    
p()

```
